```python
# One-Hot Encoding
import numpy as np

# Feature 2: User Device
devices = ["Android", "iOS", "iOS", "Android", "Web"]
d = set(devices)
dev = list(d)
state_map = {}
for i, j in enumerate(dev):
  state_map[j] = i

print(state_map)

encoded_matrix = []
for i in devices:
  x = np.zeros(3)
  k = state_map[i]
  x[k] = 1
  encoded_matrix.append(x)

print(encoded_matrix)
```

```python
#Logistic Regression
import numpy as np

# Raw z-scores (linear outputs) for 5 users
z_scores = np.array([-5.0, -1.2, 0.0, 1.5, 4.0])

def sigmoid(z):
  sigma = 1/(1 + np.exp(-z))
  return sigma

print(sigmoid(z_scores))
```

```python
#Confusion-Matrix
import numpy as np

# 1 = Fraud, 0 = Legitimate
y_true = np.array([1, 0, 1, 1, 0, 1, 0, 0])
y_pred = np.array([1, 0, 0, 1, 0, 1, 1, 0])
def manual_confusion_matrix(y_true, y_pred):
  TP = np.sum((y_true == 1) & (y_pred == 1))
  TN = np.sum((y_true == 0) & (y_pred == 0))
  FP = np.sum((y_true == 0) & (y_pred == 1))
  FN = np.sum((y_true == 1) & (y_pred == 0))
  x = np.array([[TP, FP], [FN, TP]])/len(y_true)
  return x

print(manual_confusion_matrix(y_true, y_pred))
```

```python
#Decision Trees
import numpy as np

node_A = np.array([1, 1, 1, 1, 1])       # Pure state
node_B = np.array([1, 0, 1, 0, 1, 0])    # High entropy state
node_C = np.array([1, 1, 1, 0])          # Mixed state

def calculate_gini(node):
  k = len(node)
  p1 = np.mean(node) #The Fraction of 1s
  p0 = 1 - p1
  G = 1 - p1**2 - p0**2
  return G

print(calculate_gini(node_A), calculate_gini(node_B), calculate_gini(node_C))
```
